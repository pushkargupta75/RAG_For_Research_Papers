## Local Deployment

### Step 1: Environment Setup

````python
# Check if required packages are installed
import sys
import subprocess

def check_package(package_name):
    try:
        __import__(package_name)
        return True
    except ImportError:
        return False

packages = ['streamlit', 'langchain', 'chromadb', 'openai']
missing = [pkg for pkg in packages if not check_package(pkg)]

if missing:
    print(f"⚠️ Missing packages: {', '.join(missing)}")
    print("\nInstall with: pip install -r ../requirements.txt")
else:
    print("✓ All required packages installed")
````

### Step 2: Set Environment Variables

Create a `.env` file in the project root:

````python
# Display example .env content
env_content = """
# Groq Configuration (Fast & Free LLM)
GROQ_API_KEY=your_groq_api_key_here

# Google Configuration (Free Embeddings)
GOOGLE_API_KEY=your_google_api_key_here

# Get free keys at:
# Groq: https://console.groq.com
# Google: https://makersuite.google.com/app/apikey
"""

print(".env file should contain:")
print(env_content)
````

### Step 3: Run the App Locally

````python
# Terminal command to run Streamlit
# Run this in your terminal:
# 
# streamlit run src/app.py
#
# Or with specific port:
# streamlit run src/app.py --server.port 8501

print("Run in terminal:")
print("  cd ../")
print("  streamlit run src/app.py")
print("\nThe app will open at: http://localhost:8501")
````

## Docker Deployment

### Build Docker Image

````python
# Docker commands
docker_commands = """
# Build the image
docker build -t sci_synth:latest .

# Run the container
docker run -d \\
  -p 8501:8501 \\
  -e GROQ_API_KEY=your_groq_key \\
  -e GOOGLE_API_KEY=your_google_key \\
  --name sci_synth_app \\
  sci_synth:latest

# Check logs
docker logs sci_synth_app

# Stop the container
docker stop sci_synth_app

# Remove the container
docker rm sci_synth_app
"""

print("Docker Deployment Commands:")
print(docker_commands)
````

## Hugging Face Spaces Deployment

### Method 1: Web Interface

1. **Go to Hugging Face Spaces**: https://huggingface.co/spaces
2. **Create New Space**:
   - Click "Create new Space"
   - Choose "Streamlit" as SDK
   - Name it `sci_synth`
   - Set to Public or Private

3. **Upload Files**:
   - Upload all files from `src/`
   - Upload `requirements.txt`
   - Rename `src/app.py` to `app.py` (root level)

4. **Configure Secrets**:
   - Go to Settings → Repository secrets
   - Add `OPENAI_API_KEY` or `GOOGLE_API_KEY`

5. **Deploy**:
   - Space will auto-build and deploy
   - Access at `https://huggingface.co/spaces/YOUR_USERNAME/sci_synth`

### Method 2: Git Push

````python
hf_commands = """
# Clone your space
git clone https://huggingface.co/spaces/YOUR_USERNAME/sci_synth
cd sci_synth

# Copy project files
cp -r /path/to/sci_synth/src/* .
cp /path/to/sci_synth/requirements.txt .

# Create README.md for Hugging Face
cat > README.md << 'EOF'
---
title: sci_synth
emoji: 🔬
colorFrom: blue
colorTo: green
sdk: streamlit
sdk_version: 1.31.0
app_file: app.py
pinned: false
---

# sci_synth - Scientific Paper RAG System

[Add your description here]
EOF

# Commit and push
git add .
git commit -m "Initial deployment"
git push
"""

print("Hugging Face Git Deployment:")
print(hf_commands)
````

## AWS Deployment (EC2)

### Setup EC2 Instance

````python
aws_setup = """
# 1. Launch EC2 Instance
#    - AMI: Amazon Linux 2 or Ubuntu
#    - Type: t2.medium (minimum)
#    - Security Group: Allow port 80/443

# 2. SSH into instance
ssh -i your-key.pem ec2-user@your-instance-ip

# 3. Install dependencies
sudo yum update -y
sudo yum install docker -y
sudo service docker start
sudo usermod -a -G docker ec2-user

# 4. Pull and run container
docker pull yourusername/sci_synth:latest
docker run -d \\
  -p 80:8501 \\
  -e OPENAI_API_KEY=$OPENAI_API_KEY \\
  --restart unless-stopped \\
  sci_synth:latest

# 5. Access via instance public IP
# http://your-instance-ip
"""

print("AWS EC2 Deployment:")
print(aws_setup)
````

## Google Cloud Run Deployment

````python
gcp_commands = """
# 1. Build and push to Google Container Registry
gcloud builds submit --tag gcr.io/PROJECT_ID/sci_synth

# 2. Deploy to Cloud Run
gcloud run deploy sci_synth \\
  --image gcr.io/PROJECT_ID/sci_synth \\
  --platform managed \\
  --region us-central1 \\
  --allow-unauthenticated \\
  --set-env-vars OPENAI_API_KEY=$OPENAI_API_KEY

# 3. Service will be available at generated URL
# https://sci-synth-xxxxx-uc.a.run.app
"""

print("Google Cloud Run Deployment:")
print(gcp_commands)
````

## Azure Container Instances

````python
azure_commands = """
# 1. Build and push to Azure Container Registry
az acr build --registry YOUR_ACR --image sci_synth:latest .

# 2. Deploy to Container Instances
az container create \\
  --resource-group YOUR_RG \\
  --name sci-synth \\
  --image YOUR_ACR.azurecr.io/sci_synth:latest \\
  --dns-name-label sci-synth \\
  --ports 8501 \\
  --environment-variables OPENAI_API_KEY=$OPENAI_API_KEY

# 3. Access via FQDN
# http://sci-synth.YOUR_REGION.azurecontainer.io:8501
"""

print("Azure Deployment:")
print(azure_commands)
````

## Streamlit Cloud (Easiest)

### Steps:

1. **Push to GitHub**:
   ```bash
   git init
   git add .
   git commit -m "Initial commit"
   git remote add origin https://github.com/YOUR_USERNAME/sci_synth.git
   git push -u origin main
   ```

2. **Go to Streamlit Cloud**: https://streamlit.io/cloud

3. **New App**:
   - Connect GitHub account
   - Select repository: `YOUR_USERNAME/sci_synth`
   - Main file path: `src/app.py`
   - Advanced settings → Secrets → Add API keys

4. **Deploy**:
   - Click "Deploy"
   - App will be live at `https://share.streamlit.io/YOUR_USERNAME/sci_synth`

## Testing the Deployment

### Health Check Endpoints

````python
import requests

def test_deployment(url):
    """Test if deployment is accessible"""
    try:
        response = requests.get(url, timeout=10)
        if response.status_code == 200:
            print(f"✓ Deployment accessible at {url}")
            return True
        else:
            print(f"⚠️ Got status code: {response.status_code}")
            return False
    except Exception as e:
        print(f"✗ Error accessing {url}: {str(e)}")
        return False

# Test local deployment
print("\nTesting local deployment:")
test_deployment("http://localhost:8501")

print("\nReplace with your deployed URL to test cloud deployment")
````

## Troubleshooting

### Common Issues and Solutions

**Issue 1: "No module named 'streamlit'"**
- Solution: `pip install -r requirements.txt`

**Issue 2: "Vector store not found"**
- Solution: Run notebooks 01-02 to create embeddings first

**Issue 3: "API key not found"**
- Solution: Set environment variables or add to Streamlit secrets

**Issue 4: "Out of memory"**
- Solution: Reduce chunk size, batch size, or upgrade instance

**Issue 5: "Slow response times"**
- Solution: Enable caching, use faster models, optimize queries

**Issue 6: Docker container won't start**
- Solution: Check logs with `docker logs container_name`
